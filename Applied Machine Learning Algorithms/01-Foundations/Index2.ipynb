```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
titanic = pd.read_csv('../Exercise Files/titanic.csv')
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'].fillna(titanic['Age'].mean(), inplace=True)
titanic.head(10)
```

```python
for i, col in enumerate(['SibSp', 'Parch']):
    plt.figure(i)
    sns.catplot(x=col, y='Survived', data=titanic,
                kind='point', aspect=2)
```

```python
titanic['Family_cnt'] = titanic['SibSp'] + titanic['Parch']
```

```python
titanic.head()
```

```python
titanic.drop(['PassengerId', 'SibSp', 'Parch'], axis=1,
             inplace=True)
```

```python
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
titanic.groupby(titanic['Cabin'].isnull())['Survived'].agg('mean')
```

```python
titanic[titanic['Cabin'].isnull()].shape
```

```python
titanic['Cabin'].head()
```

```python
titanic['Cabin_ind'] = np.where(titanic['Cabin'].isnull(), 0, 1)
```

```python
titanic.head()
```

```python
gender_num = {'male': 0, 'female': 1}
```

```python
titanic['Sex'] = titanic['Sex'].map(gender_num)
```

```python
titanic.head()
```

```python
titanic.drop(['Cabin', 'Embarked', 'Name', 'Ticket'], axis=1,
             inplace=True)
```

```python
titanic.to_csv('titanic_cleaned.csv', index=False)
```
